```python
def one_hot_encoding(y, num_labels):
    y_enc = np.zeros((num_labels, y.shape[1]))
    
    for i, val in enumerate(y[0]):
        y_enc[val, i] = 1.0
        
    return y_enc
```

```python
Y = np.random.permutation(10).reshape((1, 10))
print(one_hot_encoding(Y, 10))
```

```python
def relu(z):
    return np.maximum(0, z), z
```

```python
def relu_backward(dA, cache):
    Z = cache
    
    dZ = np.array(dA, copy=True)
    dZ[Z <= 0] = 0
    
    return dZ
```

```python
def softmax(z):
    exps = np.exp(z - np.max(z, axis=1, keepdims=True))
    return exps/np.sum(exps, axis=1, keepdims=True), z
```

```python
def initialize_parameters(layer_dims):
    parameters = {}
    L = len(layer_dims)
    
    for l in range(1, L):
        parameters["W" + str(l)] = np.random.randn(layer_dims[l], layer_dims[l-1]) # * np.sqrt(2 / (layer_dims[l-1] + layer_dims[l]))
        parameters["b" + str(l)] = np.ones((layer_dims[l], 1))
        
    return parameters
    
```

```python
params = initialize_parameters([X_train.shape[0], 20, 20, 10])
print(params["W1"].shape)
print(params["W2"].shape)
print(params["W3"].shape)
```

```python
def initialize_adam(parameters):
    L = len(parameters) // 2
    v = {}
    s = {}
    
    for l in range(L):
        v["dW" + str(l+1)] = np.zeros(parameters["W" + str(l+1)].shape)
        v["db" + str(l+1)] = np.zeros(parameters["b" + str(l+1)].shape)
        s["dW" + str(l+1)] = np.zeros(parameters["W" + str(l+1)].shape)
        s["db" + str(l+1)] = np.zeros(parameters["b" + str(l+1)].shape)
    
    return v, s
    
```

```python
v, s = initialize_adam(params)
print(v.keys(), s.keys())
```

```python
def random_mini_batches(X, Y, mini_batch_size=64):
    m = X.shape[1]
    minibatches = []
    
    # Shuffle (X, Y)
    permutation = list(np.random.permutation(m))
    shuffled_X = X[:, permutation]
    shuffled_Y = Y[:, permutation].reshape((1, m))
    
    # Partition shuffled_X and shuffled_Y
    num_complete_batches = math.floor(m / mini_batch_size)
    for k in range(num_complete_batches):
        mini_batch_X = shuffled_X[:, k*mini_batch_size:(k+1)*mini_batch_size]
        mini_batch_Y = shuffled_Y[:, k*mini_batch_size:(k+1)*mini_batch_size]
        
        minibatch = (mini_batch_X, mini_batch_Y)
        minibatches.append(minibatch)
    
    # Fix the last probable incomplete minibatch
    if m % mini_batch_size != 0:
        mini_batch_X = shuffled_X[:, mini_batch_size * num_complete_batches:]
        mini_batch_Y = shuffled_Y[:, mini_batch_size * num_complete_batches:]
        
        minibatch = (mini_batch_X, mini_batch_Y)
        minibatches.append(minibatch)
        
    return minibatches
```

```python
rand_mini_X, rand_mini_Y = X_train.copy(), Y_train.copy()
mini = random_mini_batches(rand_mini_X, rand_mini_Y)
print(mini[0][0].shape, mini[0][1].shape)
```

```python
def linear_forward_unit(A, W, b):
    Z = np.dot(W, A) + b
    cache = (A, W, b)
    return Z, cache
```

```python
Z, cache = linear_forward_unit(mini[0][0], params["W1"], params["b1"])
```

```python
def linear_activation_unit(A_prev, W, b, activation):
    if activation == "relu":
        Z, linear_cache = linear_forward_unit(A_prev, W, b)
        A, activation_cache = relu(Z)
    
    elif activation == "softmax":
        Z, linear_cache = linear_forward_unit(A_prev, W, b)
        A, activation_cache = softmax(Z)
        
    cache = (linear_cache, activation_cache)
    
    return A, cache
```

```python
A, cache = linear_activation_unit(mini[0][0], params["W1"], params["b1"], "relu")
```

```python
def feed_forward_propagation(X, parameters):
    L = len(parameters) // 2
    A = X
    caches = []
    
    for l in range(1, L):
        A_prev = A
        A, cache = linear_activation_unit(A_prev, parameters["W" + str(l)], parameters["b" + str(l)], "relu")
        caches.append(cache)
    
    AL, cache = linear_activation_unit(A, parameters["W" + str(L)], parameters["b" + str(L)], "softmax")
    caches.append(cache)
    
    return AL, caches
```

```python
AL, caches = feed_forward_propagation(mini[0][0], params)
```

```python
def compute_cost(AL, Y):
    m = Y.shape[1]
    
    AL = np.clip(AL, 1e-10, 1)  # limit values to 1e-8 to 1
    logpreds = Y * np.log(AL)
    error = (-1/m) * np.sum(logpreds)
    
    return error
```

```python
y_enc = one_hot_encoding(mini[1][1], 10)
print(compute_cost(AL, y_enc))
```

```python
def linear_backward_unit(dZ, linear_cache):
    A_prev, W, b = linear_cache
    m = A_prev.shape[1]
    
    dW = (1/m) * np.dot(dZ, A_prev.T)
    db = (1/m) * np.sum(dZ, axis=1, keepdims=True)
    dA_prev = np.dot(W.T, dZ)
    
    return dA_prev, dW, db
```

```python
def linear_backward_activation(dA, caches, activation):
    linear_cache, activation_cache = caches
    
    if activation == "relu":
        dZ = relu_backward(dA, activation_cache)
        dA_prev, dW, db = linear_backward_unit(dZ, linear_cache)
        
    return dA_prev, dW, db
    
```

```python
def backward_propagation(AL, Y, caches):
    grads = {}
    L = len(caches)
    m = AL.shape[1]
    
    # Initialize Backprop using softmax derivative
    dZL = AL - Y
    
    # Lth layer (SOFTMAX -> LINEAR)
    current_cache = caches[L-1]
    grads["dA" + str(L-1)], grads["dW" + str(L)], grads["db" + str(L)] = linear_backward_unit(dZL, current_cache[0])
    
    # For l = L-2...0
    for l in reversed(range(L-1)):
        current_cache = caches[l]
        dA_prev_temp, dW_temp, db_temp = linear_backward_activation(grads["dA" + str(l+1)], current_cache, "relu")
        grads["dA" + str(l)] = dA_prev_temp
        grads["dW" + str(l+1)] = dW_temp
        grads["db" + str(l+1)] = db_temp
    
    return grads
    
```

```python
grads = backward_propagation(AL, y_enc, caches)
```

```python
def update_parameters_adam(parameters, grads, v, s, t, learning_rate=0.01, beta1=0.9, beta2=0.999, epsilon=1e-8):
    L = len(parameters) // 2
    v_corrected = {}
    s_corrected = {}
    
    for l in range(L):
        v["dW" + str(l+1)] = beta1 * v["dW" + str(l+1)] + (1 - beta1) * grads["dW" + str(l+1)]
        v["db" + str(l+1)] = beta1 * v["db" + str(l+1)] + (1 - beta1) * grads["db" + str(l+1)]
        
        v_corrected["dW" + str(l+1)] = v["dW" + str(l+1)] / (1 - beta1**t)
        v_corrected["db" + str(l+1)] = v["db" + str(l+1)] / (1 - beta1**t)
        
        s["dW" + str(l+1)] = beta2 * s["dW" + str(l+1)] + (1 - beta2) * grads["dW" + str(l+1)]**2
        s["db" + str(l+1)] = beta2 * s["db" + str(l+1)] + (1 - beta2) * grads["db" + str(l+1)]**2
        
        s_corrected["dW" + str(l+1)] = s["dW" + str(l+1)] / (1 - beta2**t)
        s_corrected["db" + str(l+1)] = s["db" + str(l+1)] / (1 - beta2**t)
        
        parameters["W" + str(l+1)] = parameters["W" + str(l+1)] - learning_rate * v_corrected["dW" + str(l+1)] / (np.sqrt(s_corrected["dW" + str(l+1)]) + epsilon)
        parameters["b" + str(l+1)] = parameters["b" + str(l+1)] - learning_rate * v_corrected["db" + str(l+1)] / (np.sqrt(s_corrected["db" + str(l+1)]) + epsilon)
    
    return parameters, v, s
```

```python
v, s = initialize_adam(params)
params, v, s = update_parameters_adam(params, grads, v, s, 2)
```

```python
def model(X, Y, layer_dims, learning_rate=0.01, mini_batch_size=64, beta1=0.9, beta2=0.999, epsilon=1e-8, 
          num_epochs=1000, print_cost=True):
    L = len(layer_dims)
    costs = []
    t = 0
    m = X.shape[1]

    parameters = initialize_parameters(layer_dims)
    v, s = initialize_adam(parameters)
    
    for i in range(num_epochs):
        minibatches = random_mini_batches(X, Y, mini_batch_size)
        cost_total = 0
        
        for minibatch in minibatches:
            mini_batch_X, mini_batch_Y = minibatch
            
            mini_batch_Y = one_hot_encoding(mini_batch_Y, 10)
            
            # Forward propagation
            AL, caches = feed_forward_propagation(mini_batch_X, parameters)
            
            # Compute cost
            cost_total += compute_cost(AL, mini_batch_Y)
            
            # Backward propagation
            grads = backward_propagation(AL, mini_batch_Y, caches)
            
            # Update Parameters
            t = t + 1
            parameters, v, s = update_parameters_adam(parameters, grads, v, s, t, learning_rate, beta1, beta2, epsilon)
            
        cost_avg = cost_total / m
        
        if print_cost and i % 100 == 0:
            print(f'Cost at iteration {i}: {cost_avg}')
        if print_cost and i % 100 == 0:
            costs.append(cost_avg)
            
    plt.plot(costs)
    plt.ylabel("Cost")
    plt.xlabel("Iteration (per 100 epoch)")
    plt.title(f"Learning rate: {learning_rate}")
    plt.show()
    
    return parameters
    
```

```python
parameters = model(X_train, Y_train, [X_train.shape[0], 20, 20, 10], learning_rate=0.00067)
```
